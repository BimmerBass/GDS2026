```python
import pandas as pd
from sklearn.metrics import f1_score
import joblib
from pathlib import Path
```

```python
DATA_DIR = Path("data/preprocessed")
SPLIT_DIR = DATA_DIR / "tokenized_splits"
```

```python
model = joblib.load("data/models/advanced_model.joblib")
vectorizer = joblib.load("data/models/advanced_vectorizer.joblib")
```

```python
test_df = pd.read_csv(SPLIT_DIR / "test.csv")
y_test = test_df["is_reliable"]
y_pred = model.predict(vectorizer.transform(test_df["tokens"]))
```

```python
print(f1_score(y_test, y_pred))
```
