```python
import pandas as pd
from pathlib import Path
import joblib
from tqdm import tqdm
tqdm.pandas()
```

```python
DATA_DIR = Path("data/preprocessed")
TOKENIZER_DIR = DATA_DIR / "tokenizers"
SPLIT_DIR = DATA_DIR / "splits"
TOKEN_SPLIT_DIR = DATA_DIR / "tokenized_splits"
```

```python
tokenizer = joblib.load(TOKENIZER_DIR / "top10k.joblib")
```

```python
for split in ["train", "val", "test"]:
    df = pd.read_csv(SPLIT_DIR / f"{split}.csv", usecols=['content', 'is_reliable'])
    pd.DataFrame({
        'tokens' : df['content'].progress_map(tokenizer.text_to_ids),
        'is_reliable' : df['is_reliable']
    }).to_csv(TOKEN_SPLIT_DIR / f"{split}.csv")
```
